## <a name="Summary"></a>1. Summary

Bellabeat is a high-tech company specializing in women's wellbeing, offering smart devices to measure and optimize women's health. They measure key aspects of women’s health such as heart rate, calorie expenditure, steps, activity, intensity, and sleep. Bellabeat believes competitor data (i.e. Fitbit) can be leveraged to understand how customers use non-Bellabeat products and discover opportunities in the market to ultimately propel Bellabeat out of the "small business status" to become a leading provider in the industry.

### Data Cleaning & Manipulation
- All spreadsheet locale settings were converted to 'United States' for the spreadsheet to interpret time-related columns correctly. Settings were changed via Files > Settings > Locale = 'United States'
- Any duplicates were removed using the 'Remove Duplicates' tool. The sleepDay_merged file consisted of 3 duplicates.
- Trimmed and leading or trailing spaces 'trim whitespace' tool. No excess spaces were found.
- Timestamps were split using "text to columns" for more granular analysis in BigQuery. The time component was renamed 'ActivityHour' for all files.
- A 'DayOfWeek' column was added, using the formula 'TEXT' formatted as 'DDDD'.
- Standardized format for consistency and compatibility with BigQuery. Dates were converted to MM/DD/YYYY and time to 24hr format.
- The number of unique Ids was checked using the formula 'COUNTUNIQUE(A2:A)'
- The number of distinct days was checked using the formula 'COUNTUNIQUE(B2:B)', to ensure none exceeded 31 days
- For the dailyActivity_merged dataset, using a pivot table the total minutes per day for each Id were calculated to ensure no minutes exceeded 1440 (i.e. the amount of minutes in a day). No recorded minutes exceeded 1440 meaning there were no mis-entries for daily minutes.
- Google Sheet's data cleaning suggestion tool provided final cleaning recommendations and a quick summary of the data, via Data > Data Clean-up > Clean-up Suggestions
- For sleepDay_merged data, 3 users (Id No. 2320127002, 7007744171, 4558609924) averaged less than 4 hours sleep per night. This could indicate the users removed their devices during sleep. Upon further inspection, it was found that user 2320127002 had only 1 entry, 7007744171 had 2 and 4558609924 had 5 entries. These users were removed and reduced the total number from 24 to 21 users.
- All processed files were downloaded as csv files for local upload to BigQuery
- The weightLogInfo_merged data only contains weight information about 8 users and is an insufficient sample size to gain insights from  
<br>

### Importing to BigQuery
BigQuery automatically detected the correct schema for the columns of each spreadsheet.
The DailyActivity, Sleep, and Intensity spreadsheets were joined together:

```sql
SELECT A.*, C.DailyAvgIntensity
FROM `coursera-practice-377902.Smart_Device_Data.ActivityAndSleep_Merged` A

LEFT JOIN
(
  SELECT
    Id,
    ActivityDay,
    ROUND(AVG(TotalIntensity), 2) AS DailyAvgIntensity
  FROM `coursera-practice-377902.Smart_Device_Data.hourlyIntensities_merged`
  GROUP BY Id, ActivityDay
) C
ON A.id = C.Id AND A.ActivityDay = C.ActivityDay;
```

### 5C. Differences by user

Categorizing and comparing how users' data differ from one another and the US averages.

#### i. Average Steps

Approximately 10,000 steps per day is the recommended amount to be considered healthy. A study published by the 'Journal of American Medical Association' indicates that 10,000 steps per day significantly reduce cardiovascular disease, degenerative brain disease, and 13 types of cancer. The study noted that fewer than 10,000 steps are still associated with a great decline in health risks; however, Americans on average walk less than 4,700 steps per day. Users' steps will be categorized as: < 4,500 steps = inactive, 4,500-7,499 = light, 7,500-10,999 = moderately active, > 10,999 = very active.


<br>

#### ii. Average Minutes of Activity

The U.S. Department of Health & Human Services recommends either 150 mins of moderate-intensity aerobic exercise, 75 minutes of vigorous-intensity aerobic exercise, or a combination of both every week. This equates to 21.4 mins of moderate aerobic exercise per day or 10.7 mins of vigorous intensity per day. The aerobic exercise should be accompanied by two sessions of strength training a week. The Fitbit data provided has limited context and is difficult to interpret how minutes of activity at specific activity levels (i.e. sedentary, light, fairly active, and very active) are measured. Therefore, the Fitbit data cannot specifically be categorized as moderate and vigorous aerobic exercise. It's important to mention that regular daily movement associated with work and commuting typically does not contribute to the recommended aerobic exercise per day.  

#### iii. Average Calories

Women on average expend between 1,600 and 2,400 calories per day. Factors such as height, weight, and physical exercise affect calories burned per day. To lose weight, more calories need to be burned than calories consumed, and the opposite needs to occur to gain weight. Exercise and a balanced diet, meaning a particular selection of healthy foods, can help achieve a caloric deficit. Fitbit users burned 2,300 calories per day, which is at the higher end of average calories burned by women.  

#### iv. Average Sleep Minutes

The National Library of Medicine recommends a minimum of 420 minutes (7 hours) of sleep and no more than 640 minutes (9 hours). Gallup polls show that on average, Americans sleep 6.8 hours a night, and close to 40% of adults are not getting the recommended amount of sleep. Regular daily sleep of less than 7 hours is associated with various health risks, whereas 9 hours of sleep could be excessive. Results were categorized as follows: < 420 minutes of sleep = insufficient sleep, 420 - 540 = healthy sleep, > 540 = excessive sleep.


<br>

Another important variable is sleep quality. This includes discrepancies between time in bed and the amount of sleep, the number of REM cycles, and any other forms of sleep disturbances. On average, falling asleep should take no more than 15-20 mins. Minimal time spent awake in bed would be anything less than 15 mins; the average is between 15-20 mins; anything between 20-30 mins is mildly excessive, and according to the sleep foundation, if an individual has not fallen asleep within 30 mins, they should get up and do something relaxing. This will be categorized as excessive.


<br>

Results show that the majority of Fitbit users were awake in bed for an excessive amount of time. 76.2% were in bed awake for greater than 20 mins, and almost half were awake in bed for more than 30 mins. Less than 25% of users were awake in bed for less than 20 mins.


### 5D. Basic Trends and Correlations
Regression analysis is a statistical method to determine the relationship between a dependent variable and one or more independent variables. It reveals how changes in an independent variable will impact the dependent variable. 


### i. Sleep and Steps


<br>

There is little correlation between total steps and the amount of sleep users attained on a daily basis. With an R<sup>2</sup> value of 0.027, the correlation is not statistically significant.  
<br>
### ii. Sleep and Sedentary Minutes


<br>

The correlation between sleep and sedentary minutes is stronger with an R<sup>2</sup> value of 0.408. This makes sense as sleep is a sedentary activity, and a large percentage of users recorded sedentary activity would have been during their sleep. The amount of time sedentary can partially explain the variation in sleep minutes.  
<br>
### iii. Sleep and Calories


<br>

The correlation between sleep and calories burned is not significant with an R<sup>2</sup> value of 0.002.  
<br>
### iv. Calories Burned and Total Steps


<br>

There is a low correlation between total steps and the number of calories burned, with an R<sup>2</sup> value of 0.35. The more steps an individual takes per day correlates with more calories burned. This is possibly a low correlation as Everyday Health Inc states that the majority of calories are burned from regular life, and factors such as age and body composition are influencing factors also.  
<br>
### v. Calories Burned vs Recommended Active Minutes


<br>

According to the U.S. Department of Health & Human Services, exercise needs to be of moderate or vigorous intensity to be counted as exercise that reaps health benefits. Fairly and very active minutes meet this criteria. There is a low correlation between the recommended active minutes and calories burned. With an R<sup>2</sup> value of 0.287, it is not statistically significant, as the recommended active minutes do not explain the majority of the variance in calories burned.  
<br>
### vi. Daily Average Intensity and Calories


<br>

There is a low correlation between Intensity and Calories burned. With an R<sup>2</sup> value of 0.357, calories burned is partially explained by users' daily average intensity. The greater the daily intensity score, the more calories burned per day. Variables such as users' default rate of calories burned without exercise, body composition, and age would most likely explain more of the variance in daily calories burned.  

### 5E. Trends by Time/ Day

#### i. Average Active Minutes and Average Steps by Day of Week
The average active minutes were quite consistent and had small variation across the week, scoring between approximately 208 and 245 active minutes per day. Saturday had the most active minutes (244.27 mins) and Sunday had the lowest active minutes (208.49 mins). <br>
#### ii. Sleep per Day of Week
In accordance with the U.S. Department of Health & Human Services, Saturday and Sunday met the recommended sleep. This is possibly due to people being less busy on the weekends. However, from Monday to Friday, users had insufficient sleep, averaging 410 minutes during workdays, which is on average 10 minutes less than the recommended.  
<br>
#### iii. Average Calories Burned by Day of Week
The heatmap shows that Tuesday and Saturday recorded the highest amount of calories burned, which is in accordance with the number of active minutes per day that also peaked on Tuesday and Saturday. Calorie expenditure is more widespread across Monday to Friday, as calorie expenditure starts to increase at 5 am. Comparatively, significant calorie expenditure occurred over fewer hours on the weekend. Calories burned on Saturday and Sunday started to increase at 7 am. Calories burned on average peaked around 5-8 pm, possibly due to it being a popular time to exercise after work.  
<br>
#### iv. Average Heartrate by Time & Day of Week
In comparison to the calories burned heatmap, heart rate records were a bit more sporadic across the time and day of the week. Average heart rates had small variation across each day of the week. Heart rates on average were highest on Fridays at 79.01 BPM and were lowest on Sundays at 73.97 BPM. Additionally, there are slightly higher heart rate figures in the early mornings of Monday and Saturday. Between 4 pm and 9 pm, heart rates were consistently high, scoring above 81 BPM for each hour during this timeframe.  
<br>
#### vi. Average Intensity by Time and Day of Week
Average intensity had a wide amount of variation for each day of the week. Surprisingly, the average intensity was significantly higher on Tuesday compared to any other day of the week. Users' average Intensity was lowest on Sunday at 530. Similar to average calorie expenditure, intensity scores are more widespread across Monday to Friday, as Intensity starts to increase at 5 am and decreases rapidly at 11 pm. The timeframe with the highest intensities was from 5 pm to 8 pm, averaging over 320 per hour.
